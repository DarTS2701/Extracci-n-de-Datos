# **FILTROS VARIABLES CUALITATIVAS**

```python
#Filtro por objeto
#Variables: Nombre
filtro1=df[df["name"] == "Mexico"]
filtro1
```

```python
#Filtro por filas
#Variable:Filas.

filtro2=df.iloc[[3,7,100] , : ]  #filas no consecutivas
filtro2
```

```python
#Filtro por columnas
#Variable: columnas
filtro3=df.iloc[ : , 3:7] #Columnas consecutivas
filtro3
```

```python
# Paso 1 filtro 4 Leer denuevo en df

df= pd.read_csv("CountryTable.csv", index_col=2)
df
```

```python
#Paso 2: Aplicamos Filtro de filas vs columnas
#Variable: Filas y Columnas

filtro4=df.loc[["Asia","Africa"], ["local_name"]]
filtro4
```

```python
#Filtro por cabecera
#Variable: Cabecera
filtro5=df.head(7)
filtro5
```

# **Filtro variables cuantitativas**

```python
#Filtro por comparacion
#Variable: Capital
filtro6=df[df["capital"] > 2000]
filtro6
```

# **Filtros de conectores logicos**

```python
#Filtro AND
#Variables: gnp y government_form
filtro7=df[(df["gnp"] > 7000) & (df["government_form"] == "Republic")]
filtro7
```

```python
#Filtro or
#Variables: government_form y life_expectancy
filtro8=df[(df["life_expectancy"] > 80) | (df["government_form"] == "Federation")]
filtro8
```

```python
#Filtro not
#Variable: Capital y surface_area
filtro9=df[~(df["capital"] >= 10000) & ~(df["surface_area"] >= 10000)]
filtro9
```
